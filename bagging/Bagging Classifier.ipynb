```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import seaborn as sns
```

```python
x,y = make_moons(n_samples=1000,noise=0.2)
```

```python
df = {'data1':x[:,0],'data2':x[:,1],'output':y}
```

```python
db = pd.DataFrame(df)
```

```python
sns.scatterplot(x='data1',y='data2',data=db,hue='output')
plt.show()
```

```python
x = db.iloc[:,:-1]
y = db['output']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
bg = BaggingClassifier(estimator=SVC(),n_estimators=30)
bg.fit(x_train,y_train)
bg.score(x_train,y_train)*100,bg.score(x_test,y_test)*100
```

```python
svc = SVC()
svc.fit(x_train,y_train)
svc.score(x_train,y_train)*100,svc.score(x_test,y_test)*100
```

```python
rf = RandomForestClassifier(n_estimators=30)
rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train)*100,rf.score(x_test,y_test)*100
```
